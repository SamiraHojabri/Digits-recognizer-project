```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import csv
```

```python
data_train = pd.read_csv("train.csv")
data_train
```

```python
Dtc = DecisionTreeClassifier()
```

```python
# This section trains the model over the TrainSet (train.csv); The dataset of 42000 pictures in the file train.csv are equally divided into two sets:
# 1) data_train to train the model including the first 21000 labeled pictures; 2) data_test to test the precision of the trained model over the last 21000 lebeled pictures
x_train = data_train.iloc[0:21000,1:]
y_train = data_train.iloc[0:21000,0]

# The Decision Tree Classifier is used to solve the problem and the the algorithm is trained on data_train
Dtc.fit(x_train,y_train)
```

```python
# This section tests the model over the TestSet (data_test)
x_test = data_train.iloc[21000:,1:]
y_test = data_train.iloc[21000:,0]
p = Dtc.predict(x_test)
```

```python
# The following lines calculate the accuracy of the model comparing the model's prediction against the actual labels:
count=0
for i in range(0,21000):
    count+=1 if p[i]==y_test.iloc[i] else 0
    
print("Precision = ", 100*(count/21000))    
```

```python
# Printing the output of the algorithm on the data_test in a .csv file named TrainResult.csv:
with open('TrainResult.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow([ "ImageId", "Labels"])
    for i in range(0,21000):
        writer.writerow([i+1, p[i]])
file.close()
```

```python
data_test = pd.read_csv("test.csv")
data_test
```

```python
# This section predicts the labels for the images stored in the file test.csv
x_test1 = data_test.iloc[:,:]
predict = Dtc.predict(x_test1)
predict.shape
```

```python
# Printing the output of the algorithm on the testData in a .csv file named sample_submission.csv:
with open('sample_submission.csv', 'w', newline ='') as file_predict:
    writer_predict = csv.writer(file_predict)
    writer_predict.writerow(["ImageId", "Labels"])
    for i in range(0,28000):
        writer_predict.writerow([i+1, predict[i]])
file_predict.close()
```
